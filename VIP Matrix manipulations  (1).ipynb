```python
# Define function to generate plot
def generate_plot(df, title, dates, weights=None, sharey=True, y_format=None):
    
    # PLotting parameters
    cmap = cm.tab10
    mpl.style.use("default")
    params = {
        'axes.labelsize': 12,
        'axes.titlesize': 14, 
        'legend.fontsize': 10,
        'xtick.labelsize': 12, 
        'ytick.labelsize': 12,
        'figure.figsize': (16, 4.5),
        'figure.titlesize': 14,
    }
    rcParams.update(params)
    alpha = .75
    gridspec_kw = {'width_ratios': [1, .75, 1]}
    
    # Get rid of NA
    df = df.dropna()

    # Assign trends and seasonality
    hp, sd = hpfilter(df, lamb=lamb), seasonal_decompose(df)
    df = df\
        .assign(trend = hp[1])\
        .assign(season = sd.trend + sd.seasonal)

    # Plot initializing
    f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=sharey, gridspec_kw=gridspec_kw)

    # Time-series of last 16 weeks
    df.iloc[:, [0, 1]].plot(ax=ax1)
    ax1.set_xlabel("")
    ax1.yaxis.grid(True, alpha=.25)
    ax1.legend(bbox_to_anchor=(0, 1.125), loc='upper left', ncol=2)
    if y_format:
        ax1.get_yaxis().set_major_formatter(ticker.FuncFormatter(lambda x, p: y_format.format(x)))
    for update, date in dates: 
        update_start = pd.to_datetime(date)
        if update_start in df.index:
            ax1.vlines(x=update_start, ymin=0, ymax=df.max().max(), linestyles="--", alpha=.5)

    # Barplot week comparaison
    df.iloc[-14:, [0]]\
        .assign(Week=lambda x: x.index.strftime("%Y - Week %W"))\
        .groupby("Week")\
        .agg(lambda x: np.average(x, weights=weights.loc[x.index] if weights is not None else weights))\
        .plot\
        .bar(ax=ax2, color=[cmap.colors[0:2]])
    ax2.set_ylabel("")
    ax2.set_xlabel("")
    ax2.legend().remove()
    ax2.set_xticklabels(np.unique(df.iloc[-14:, [0]].index.strftime("Week\n%W")), rotation=0)
    ax2.yaxis.grid(True, alpha=.25)
    if y_format:
        ax2.get_yaxis().set_major_formatter(ticker.FuncFormatter(lambda x, p: y_format.format(x)))

    # Barplot day comparaison
    df.iloc[-14:, [0]]\
        .assign(week=lambda x: x.index.strftime("week %W"))\
        .assign(day=lambda x: x.index.strftime("%a"))\
        .set_index("day")\
        .pivot(columns="week", values=df.columns[0])\
        .loc[["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]]\
        .plot\
        .bar(ax=ax3)
    for x, y in zip(
        np.arange(0, 7), 
        df.loc[lambda x: pd.notnull(x.season), ["season"]]\
            .assign(day=lambda x: x.index.strftime("%a"))\
            .groupby("day")\
            .agg(lambda x: np.average(x, weights=weights.loc[x.index] if weights is not None else weights))\
            .loc[["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"], "season"]
    ):
        plt.plot(
            (x-.25, x+.25), 
            (y, y),
            c="black",
            label="daily trend" if x==6 else None
        )
    ax3.yaxis.grid(True, alpha=.25)
    ax3.set_xlabel("")
    ax3.legend(bbox_to_anchor=(0, 1.125), loc='upper left', ncol=3)
    if y_format:
        ax3.get_yaxis().set_major_formatter(ticker.FuncFormatter(lambda x, p: y_format.format(x)))
    plt.suptitle(title)
        
```

```python
generate_plot(ren, "1-MONTH VIP SUBSCRIPTIONS SOLD", JurassicWorld.dates, y_format="{:,}")
```
